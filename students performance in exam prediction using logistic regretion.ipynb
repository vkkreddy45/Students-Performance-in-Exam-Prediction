```python
dataset['total']=dataset['math score']+dataset['reading score']+dataset['writing score']
```

```python
dataset['average']=dataset['total']/3
```

```python
dataset['math_pass status']=np.where(dataset['math score']<40,'F','P')
```

```python
dataset['reading_pass status']=np.where(dataset['reading score']<40,'F','P')
```

```python
dataset['writing_pass status']=np.where(dataset['writing score']<40,'F','P')
```

```python
dataset['result']=np.where(dataset['average']<50,'F','P')
```

```python
x=dataset.iloc[:,9:10].values
```

```python
y=dataset.iloc[:,13].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
classifier.predict([[8]])
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(y_test,y_pred)
```
